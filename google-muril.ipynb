```python
from transformers import pipeline
```

```python
%%time
sentiment_analysis = pipeline("sentiment-analysis", model="google/muril-base-cased")
```

```python
result = sentiment_analysis("I love you")[0]
```

```python
result
```

```python
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```
